```python
# Downloading dataset directly into google colab via opendatasets
od.download("https://www.kaggle.com/datasets/himanshunakrani/iris-dataset")
```

```python
# Importing pandas as pd to get and read csv dataset
import pandas as pd
iris_dataset = pd.read_csv("/content/iris-dataset/iris.csv")

# head() function shows the first five rows of the dataset
iris_dataset.head()
```

```python
# tail() function shows the last 5 rows of the dataset
iris_dataset.tail()
```

```python
# Using shape() function for getting the shape of the dataset
# shape() function shows number of rows and columns
iris_dataset.shape
```

```python
# Using column() function to get the cols of dataset and showing it in a DataFrame
iris_dataset_dataframe = pd.DataFrame(iris_dataset.columns)
iris_dataset_dataframe
```

```python
# Using info() and describe() to get some knowledge of the dataset
iris_dataset.info(), iris_dataset.describe()
```

```python
# Importing matplotlib and seaborn library python to visualize dataset
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Checking if null value exists
iris_dataset.isnull().sum()
```

```python
# Checking the number of species of each category and their number in the dataset
iris_dataset['species'].value_counts()
```

```python
# Create a bar graph of the dataset
iris_dataset['petal_length'].plot.bar(figsize = (20, 10));
```

```python
# Pairploting whole data via sns
sns.pairplot(iris_dataset);
```

```python
# Pairploting species of the dataset
sns.pairplot(iris_dataset,hue='species');
```

```python
# Create a scatter plot to show relationships between features.
plt.figure(figsize=(10, 8))
sns.jointplot(data=iris_dataset, x=iris_dataset['sepal_length'], y=iris_dataset['sepal_width'],
              hue='species', kind='scatter')
plt.suptitle('Sepal Length (cm) vs Sepal Width (cm)', y=1.02)
plt.grid()
plt.show()
```

```python
# Create a scatter plot to show relationships between features.
plt.figure(figsize=(10, 8))
sns.jointplot(data=iris_dataset, x=iris_dataset['petal_length'], y=iris_dataset['sepal_width'],
              hue='species', kind='scatter')
plt.suptitle('Sepal Length (cm) vs Sepal Width (cm)', y=1.02)
plt.grid()
plt.show()
```

```python
# Create a scatter plot to show relationships between features.
plt.figure(figsize=(10, 8))
sns.jointplot(data=iris_dataset, x=iris_dataset['sepal_length'], y=iris_dataset['petal_length'],
              hue='species', kind='scatter')
plt.suptitle('Sepal Length (cm) vs Sepal Width (cm)', y=1.02)
plt.grid()
plt.show()
```

```python
# Use histograms to show value distributions.
# Create histogram for 'sepal length (cm)'
sns.histplot(data=iris_dataset, x=iris_dataset['sepal_length'], bins=20)

# Show plot
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Count')
plt.show()
```

```python
# Use histograms to show value distributions.
# Create histogram for 'sepal width (cm)'
sns.histplot(data=iris_dataset, x=iris_dataset['sepal_width'], bins=20)

# Show plot
plt.title('Histogram of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Count')
plt.show()
```

```python
# Use histograms to show value distributions.
# Create histogram for 'petal_length (cm)'
sns.histplot(data=iris_dataset, x=iris_dataset['petal_length'], bins=20)

# Show plot
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Count')
plt.show()
```

```python
# Use box plots to identify outliers.
plt.figure(figsize=(10,6))
sns.boxplot(data=iris_dataset, palette='viridis')
plt.title('Box Plots: By Features')
plt.grid()
plt.show()
```
